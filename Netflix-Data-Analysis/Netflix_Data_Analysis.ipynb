```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use("seaborn-v0_8")
sns.set(font_scale=1.1)
```

```python
netflix = pd.read_csv("netflix_titles.csv")
```

```python
print("Netflix Dataset Shape:", netflix.shape)
print(netflix.head())
```

```python
netflix.info()
print(netflix.describe(include="all"))
```

```python
netflix['type'].value_counts().plot(kind='bar', color=["skyblue","orange"])
plt.title("Count: Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
netflix['listed_in'].str.split(', ', expand=True).stack().value_counts().head(10).plot(kind='barh')
plt.title("Top 10 Genres on Netflix")
plt.show()

```

```python
netflix['release_year'].value_counts().sort_index().plot()
plt.title("Netflix Releases Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```
